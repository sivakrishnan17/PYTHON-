### task 3:

```python
age=int(input("Enter your age :"))
if age<13:
    print("children")
elif 13<= age <18:
    print("Teenage")
elif 18<= age <29:
    print("Adult")
elif 29<= age <=59:
    print("Citizen")
else:
    print("Senior Citizen")
```
