### task 3:

```python
# find odd number:
list_num=[1,45,7,8,6,3,48,2]
for num in list_num:
    if num%2==0:
        continue
    else:
        print(f"{num} is odd.")
```

```python
list_num=[1,45,7,8,6,3,48,2]
for num in list_num:
    if num%2==0:
        print(f"{num} is even.")
    else:
        print(f"{num} is odd.")
```
