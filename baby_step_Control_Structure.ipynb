```python
# some people give str in int input box ,so use erroe handle method:
print(".....Check your BMI .....")
print("Press 'q' to quit")
while True:
    BMI=input("Enter the BMI :")
    if BMI.lower()=='q':
        break
    try:
        BMI=int(BMI)
        if BMI<=18:
            print("UnderWeight")
        elif 18<BMI<25:
            print("Normal")
        elif 25<= BMI <=29:
            print("OverWeight")
        else:
            print("Obesity")
    except ValueError:
        print("Invalid input try again...")



```
